Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the boston House Price Dataset

```python
house_price_dataframe = pd.read_csv("Data.csv")
```

 Understaning the correlation between varoius features in datasets

```python
 correlation  = house_price_dataframe.corr()
```

```python
#contructing a heat map to understand the correlation
plt.figure(figsize = (10,10))
sns.heatmap(correlation,cbar = True,square = True, fmt = '.1f',annot = True,annot_kws = {"size":8},cmap = "Blues")
```

```python
X = house_price_dataframe.drop(['PRICE'],axis=1)
Y = house_price_dataframe["PRICE"]
print(X)
print(Y)
```

Splitting the data

```python
X_train, X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

XGBoost Regression

```python
#Loading the model
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

Evaluation

Prediction on training data

```python
#Accuracy on prediction of trianing data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

we will now evaluate the error

Visualizing the actual prices and prediction

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual Price vs Predicted Price")
plt.show()
```

Prediction on test data

```python
#R Squared Error
score_1 = metrics.r2_score(Y_train,training_data_prediction)

#Mean Absolute error
score_2 = metrics.mean_absolute_error(Y_train,training_data_prediction)
print("R squared error : ",score_1)
print("Mean absolute Error : ",score_2)
```

```python
test_data_predicion = model.predict(X_test)
```

```python
#R Squared Error
score_1 = metrics.r2_score(Y_test,test_data_predicion)

#Mean Absolute error
score_2 = metrics.mean_absolute_error(Y_test,test_data_predicion)
print("R squared error : ",score_1)
print("Mean absolute Error : ",score_2)
```
