Importing the Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection and Analysis

PIMA Diabetes Dataset

```python
#Loading the dataset to pandas DataFrames
diabetes_dataset = pd.read_csv("diabetes.csv")
```

```python
#sepearating data and label
X = diabetes_dataset.drop(columns = 'Outcome',axis = 1)  #axis is 1 if u are dropping a column
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

Data Standardization

```python
scaler = StandardScaler() #the difference in the range of values may cause difficulty in model building so we make them all standardized                   
```

```python
scaler.fit(X) # we standardize all features in specific range
```

```python
standardized_data = scaler.transform(X) # transforms data ina particular range
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

Train Test Split

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size =0.2, stratify = Y, random_state  =2)
```

```python
print(X.shape, X_train.shape,X_test.shape)
```

Training the Model

```python
classifier = svm.SVC(kernel = 'linear') #loads support vector machine, uses linear model
```

```python
#trainig the svm classifier
classifier.fit(X_train,Y_train) 
```

Model Evaluation

Accuracy Score

```python
#Accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy Score of training data: ",training_data_accuracy)
```

```python
#Accuracy score of testing data
X_test_prediction = classifier.predict(X_test)
testing_data_accuaracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy score for testing data: ",testing_data_accuaracy)
```

If training data accuracy is very high and tesing is very low then it specifies that the model got over trained

Making a predictive system for input data

```python
input_data = (4,110,92,0,0,37.6,0.191,30)
#changing input list to numpy array
input_data_as_numpy_array = np.asarray(input_data)

#reshape the array as we are predicting for one instance
# numpy makes reshape easier
# we had the shape of our data as some (678,9) but now are only giving one data point so we have to reshape for no confusion in modell
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize input datastd
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)
if(prediction[0] ==0):
    print("The person is Not Diabetic")
else:
    print("The person is Diabetic")
```
